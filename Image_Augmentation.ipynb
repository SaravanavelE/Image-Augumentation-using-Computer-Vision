```python
cv2_imshow(img[100:250])
```

```python
img2 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(img2)
```

```python
img2 = cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(img2)
```

```python
# vertical flip
img3 = cv2.flip(img,0)
cv2_imshow(img3)
```

```python
# horizontal flip
img3 = cv2.flip(img,1)
cv2_imshow(img3)
```

```python
# resize ( compressing the image
cv2.resize(img,(250,250))
```

```python
# blurring
blurring = cv2.blur(img,(10,5))
cv2_imshow(blurring)
```

```python
# blurring
blurring = cv2.blur(img,(100,100))
cv2_imshow(blurring)
```

```python
# padding
img4 = cv2.copyMakeBorder(img,100,100,100,100,cv2.BORDER_CONSTANT)
cv2_imshow(img4)
```

```python
# braightness and contrast
img5 = cv2.convertScaleAbs(img,alpha=100,beta=1) # alpha = contrast, bete = braightness
cv2_imshow(img5)
```

```python
# Noise injuction
img6 = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,20)
cv2_imshow(img6)
```

```python
# guassian noise
img7 = cv2.GaussianBlur(img,(5,5),0)
cv2_imshow(img7)
```

```python
# color jitter
img8 = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(img8)
```

```python
# shearing
img9 = cv2.warpAffine(img,np.float32([[1,0.5,0],[0.5,1,0]]),(img.shape[1],img.shape[0]))
cv2_imshow(img9)
```

```python
# Translation
img10 = cv2.warpAffine(img, np.float32([[1, 0, 100], [0, 1, 100]]), (img.shape[1], img.shape[0]))
cv2_imshow(img10)
```

```python
# Canny
img = cv2.imread('/content/Cute_dog.jpg')
img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(img,150,80)
cv2_imshow(edges)
```
